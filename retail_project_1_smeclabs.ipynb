##### 2) Is there any correlation between customer income and their purchasing patterns?

```python
# Convert categorical income to numeric values
income_mapping = {'low': 1, 'medium': 2, 'high': 3}
retail_sales['Income'] = retail_sales['Income'].map(income_mapping)

# Calculate the correlation matrix
correlation_matrix = retail_sales[['Income', 'Total_Amount']].corr()

print("Correlation Matrix:")
print(correlation_matrix)
correlation_matrix.plot(kind='bar', title='correlation', xlabel='Income', ylabel='Total Amount')
plt.show()
```

##### 3) What are the top cities, states, or countries contributing to the transactions?

```python
# Group by City and calculate total transactions
city_totals = retail_sales.groupby('City')['Amount'].sum().sort_values(ascending=False)
print("Total Transactions by City:")
print(city_totals)

# Group by State and calculate total transactions
state_totals = retail_sales.groupby('State')['Amount'].sum().sort_values(ascending=False)
print("\nTotal Transactions by State:")
print(state_totals)


# Group by Country and calculate total transactions
country_totals = retail_sales.groupby('Country')['Amount'].sum().sort_values(ascending=False)
print("\nTotal Transactions by Country:")
print(country_totals)

```

##### 2) How do purchase trends on weekdays compare to weekends?

```python
# Extract the day of the week from PurchaseDate (0=Monday, 6=Sunday)
retail_sales['DayOfWeek'] = retail_sales['Date'].dt.dayofweek

# Classify days as weekdays (0-4) or weekends (5-6)
retail_sales['DayType'] = retail_sales['DayOfWeek'].apply(lambda x: 'Weekend' if x >= 5 else 'Weekday')

# Group by DayType and calculate the frequency and amount of purchases
purchase_trends = retail_sales.groupby('DayType').agg(
    Frequency=('Transaction_ID', 'count'),
    TotalAmount=('Amount', 'sum'),
    AverageAmount=('Amount', 'mean')
)

print("Purchase Trends by Day Type:")
print(purchase_trends)
```
